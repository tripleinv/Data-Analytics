# Groupby and aggregate function in Pandas

```python
df = pd.read_csv("https://bit.ly/3kGDuKx", na_values = "?")
df
```

```python
# groupby -  single group single column
df.groupby(["pclass"])['survived'].sum()
```

```python
# groupby - multiple columns
df.groupby(["pclass"])[['survived', 'fare']].sum()
```

```python
# groupby - multiple groups
df.groupby(["pclass","sex"])["survived"].sum()
```

```python
# groupby - multipe groups as columns
df.groupby(["pclass","sex"], as_index = False)["survived"].sum()
```

```python
# Multiple Functions
df.groupby(["pclass"])["survived"].agg(['mean','sum'])
```

```python
# set up new groups by using cut()
# df['age_bins'] = pd.cut(df['age'], bins = 3, labels = ["young","middle_age","old"])
print(df['age'].dtype)
print(df['age'].isnull().sum())

# how to handle numeric calculations with NaN values in the column: 
df["age"].fillna(df['age'].mean(), inplace = True)
```

```python
df['age_bins'] = pd.cut(df['age'], bins = 3, labels = ["young","middle_age","old"])
df.head(5)
```

# Joining and merging dataframes

```python
import pandas as pd

df1 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2001, 2002, 2003, 2004])

df2 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Int_rate':[2, 3, 2, 2],
                    'US_GDP_Thousands':[50, 55, 65, 55]},
                   index = [2005, 2006, 2007, 2008])

df3 = pd.DataFrame({'HPI':[80,85,88,85],
                    'Unemployment':[7, 8, 9, 6],
                    'Low_tier_HPI':[50, 52, 50, 53]},
                   index = [2001, 2002, 2003, 2004])

print(df1)
print(df2)
print(df3)
```

```python
print(pd.merge(df1, df2))
```

```python
# merger is better than join: to combine two data sets side by side/horizontally. 
# concat is better than append: to stack two data sets vertically. 

pd.merger(df1, df2, on = '') # this mergers the data row that has the same value on the specified'' columns, like inner join)

# suffixes & merging on columns that are unique to each DataFrame
pd.merge(df1, df2, suffixes = ['_l','_r'], left_on = "Col_2", right_on= "Col_A") # this will combine raws when values in Col_2 and Col_a matches. 

# jointing on indices
pd.merge(df1, df2, suffixes = ['_l','_r'], left_index = True, right_index = True)

## Note all the above are in default Inner joins..
# Outer_join = pd.merge(df1, df2, on = "col_name", how = "outer")

# Left_join = pd.merge(df1, df2, on = "col_name", how = "left")

# Right_join = pd.merge(df1, df2, on = "col_name", how = "right")

# Concat, the default join is outer.
# pd.concat([df1, df2], ignore_index = True) # note how you pass the df as a list, and reset the index)
# pd.concat([df1, df2], axis = 0, join = "inner")

# append()
# df1.append(df2) # vertically. 
```
